```python
def character_frequency(s):
    freq = {}

    for ch in s:
        if ch in freq:
            freq[ch] += 1
        else:
            freq[ch] = 1

    return freq


# Input
text = "algonex"

# Output
result = character_frequency(text)
print(result)
```

```python
def find_duplicates(transactions):
    freq = {}
    duplicates = []
    for transaction in transactions:
       if transaction in freq:
            freq[transaction] += 1
        else:
            freq[transaction] = 1

    for transaction, count in freq.items():
        if count > 1:
            duplicates.append(transaction)

    return duplicates
```

```python
def get_highest(employees):
    highest_name = employees[0][0]
    highest_salary = employees[0][1]

    for name, salary in employees:
        if salary > highest_salary:
            highest_salary = salary
            highest_name = name

    return highest_name


def get_lowest(employees):
    lowest_name = employees[0][0]
    lowest_salary = employees[0][1]

    for name, salary in employees:
        if salary < lowest_salary:
            lowest_salary = salary
            lowest_name = name

    return lowest_name


def get_average(employees):
    total = 0

    for name, salary in employees:
        total += salary

    return total / len(employees)


def process_employees(employees):
    highest = get_highest(employees)
    lowest = get_lowest(employees)
    average = get_average(employees)

    return highest, lowest, average


# Input
employees = [
    ("John", 50000),
    ("David", 70000),
    ("Smith", 40000),
    ("Alice", 90000)
]

# Output
highest, lowest, average = process_employees(employees)

print("Highest =", highest)
print("Lowest =", lowest)
print("Average =", int(average))
```
