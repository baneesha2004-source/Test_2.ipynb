```python
def get_grade(marks):
    if marks >= 90:
        return "A+"
    elif marks >= 80:
        return "A"
    elif marks >= 70:
        return "B"
    elif marks >= 60:
        return "C"
    elif marks >= 50:
        return "D"
    else:
        return "F"

# Test values
test_marks = [95, 82, 71, 55, 45]

# Print grades
for marks in test_marks:
    print(marks, "->", get_grade(marks))
```
