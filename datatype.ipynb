```python
student={
        "name":"aneesha",
        "cgpa":9.5,
        "location":"chennai",
        "city":"bangalore"
}
# pop()
student.pop("cgpa")
# del
del student["city"]
# clear()
student.clear()
print(student)

```

```python
# loop keys
for key in student:
    print(key)

# loop values
for value in student.values():
    print(value)

# loop items
for key, value in student.items():
    print(key, value)
```
